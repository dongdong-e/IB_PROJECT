## 강수량 데이터 정리

```python
data = pd.read_excel('제주강수량(2000-2018).xlsx', encoding='utf-8')
data.head()
```

```python
data['제주강수량(m)'] = data['제주강수량(mm)'] * 0.001
data.head()
```

```python
data['제주면적곱하기(m^3)'] = data['제주강수량(m)'] * 1849000000
data.head()
```

```python
data['리터로환산'] = data['제주면적곱하기(m^3)'] * 1000
data.head()
```

```python
import calendar 
# 366 if calendar.isleap(2003) else 365 
data['취수량추가(L)'] = 4700 * 1000 * 365
data
```

```python
df = data[(data['연도']>=2002) & (data['연도'] <= 2017)]
df
```

```python
df2 = pd.read_csv('water_usage.csv', encoding='euc-kr')
df2 = df2[::-1]
df2 = df2.rename(index=str, columns={"Unnamed: 0": "연도"})
df2
```

```python
df3 = pd.merge(df, df2, how='outer', on='연도')
df3
```

```python
df4 = df3
```

```python
df4['사용합계'] = df4['생활용수'] + df4['공업용수'] + df4['농업용수'] + df4['기타용수']
df4
```

```python
df4['지하수함량'] = df4['리터로환산']*0.41
df4
```

```python
df4.iloc[:,5][0:5] = 868 * 1000 * 365
df4.iloc[:,5][5:11] = 2100 * 1000 * 365
df4.iloc[:,5][11:] = 4600 * 1000 * 365
df4
```
